```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit
from xgboost import XGBClassifier

# Kreiranje vremenske kros validacije
tscv = TimeSeriesSplit(n_splits=3)

# Definisanje pipeline-a
pipe = Pipeline([
    ('feature_selection', SelectKBest(score_func=f_classif)),
    ('xgb', XGBClassifier(use_label_encoder=False, eval_metric='error', verbosity=0))
])

# Parametri za GridSearch
#param_grid = {
#    'feature_selection__k': [10, 20, 25],
#    'xgb__n_estimators': [500, 1000],
#    'xgb__max_depth': [5, 10],
#   'xgb__learning_rate': [0.05, 0.1]
#}

# da ne bismo vise puta radio optimizaciju
param_grid = {
    'feature_selection__k': [25],
    'xgb__learning_rate': [0.05],
    'xgb__max_depth': [5],
    'xgb__n_estimators': [1000]
}

# GridSearch sa TimeSeriesSplit
grid = GridSearchCV(pipe, param_grid, cv=tscv, scoring='accuracy', n_jobs=-1)

# Priprema podataka
X_filtered = X_train.drop(columns=['item_nbr', 'WeekStartDate', 'family'])
grid.fit(X_filtered, y_train)

# Ispis rezultata
print("Najbolji parametri:", grid.best_params_)
print("Najbolja taÄnost:", grid.best_score_)

# Prikaz izabranih feature-a
best_pipeline = grid.best_estimator_
selected_features_mask = best_pipeline.named_steps['feature_selection'].get_support()
selected_features = X_filtered.columns[selected_features_mask]
print("Odabrani feature-i:", selected_features.tolist())
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Pristup treniranom XGBoost modelu
xgb_model = best_pipeline.named_steps['xgb']

# Prikaz importansi samo za odabrane feature-e
importances = xgb_model.feature_importances_
selected_feature_names = selected_features.tolist()

# Sortiranje po vaÅ¾nosti
indices = np.argsort(importances)[::-1]
sorted_features = [selected_feature_names[i] for i in indices]
sorted_importances = importances[indices]

# Plotovanje
plt.figure(figsize=(10, 6))
plt.barh(range(len(sorted_features)), sorted_importances[::-1], align='center')
plt.yticks(range(len(sorted_features)), sorted_features[::-1])
plt.xlabel("Feature Importance")
plt.title("XGBoost Feature Importance (odabrani feature-i)")
plt.tight_layout()
plt.show()
```

```python
import optuna
from xgboost import XGBClassifier
from sklearn.pipeline import Pipeline
from sklearn.model_selection import TimeSeriesSplit, cross_val_score

# 1. Koristi veÄ‡ selektovane kolone
selected_columns = X_filtered.columns[best_pipeline.named_steps['feature_selection'].get_support()]
X_selected = X_filtered[selected_columns]

# 2. DefiniÅ¡i Optuna objective funkciju sa pipeline-om
def objective(trial):
    xgb_params = {
        'n_estimators': trial.suggest_int('n_estimators', 100, 2000),
        'max_depth': trial.suggest_int('max_depth', 3, 10),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3),
        'reg_lambda': trial.suggest_float('reg_lambda', 0.0, 1.0),
        'eval_metric': 'error',
        'verbosity': 3
    }

    # Pipeline sa samo jednim korakom â€“ XGBoost modelom
    pipe = Pipeline([
        ('xgb', XGBClassifier(**xgb_params))
    ])

    # TimeSeriesSplit
    tscv = TimeSeriesSplit(n_splits=3)

    # Evaluacija
    scores = cross_val_score(pipe, X_selected, y_train, cv=tscv, scoring='accuracy', n_jobs=-1)

    return scores.mean()

# 3. Optuna optimizacija
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=30, show_progress_bar=True)

# 4. Ispis rezultata
print("Najbolji parametri:", study.best_params)
print("Najbolja taÄnost:", study.best_value)
```

```python
# 1. Koristi veÄ‡ selektovane kolone
selected_columns = X_filtered.columns[best_pipeline.named_steps['feature_selection'].get_support()]
X_selected = X_filtered[selected_columns]


best_params = {
    'n_estimators': 288,
    'max_depth': 11,
    'min_samples_split': 7,
    'min_samples_leaf': 2,
    'max_features': 'log2'
}
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, confusion_matrix, classification_report, roc_auc_score, roc_curve
)
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Treniranje modela sa najboljim parametrima
final_model = RandomForestClassifier(**best_params)
final_model.fit(X_selected, y_train)

# 2. Priprema test skupa sa istim kolonama
X_test_filtered = X_test[selected_columns]

# 3. Predikcija
y_pred = final_model.predict(X_test_filtered)
y_proba = final_model.predict_proba(X_test_filtered)[:, 1]  # za ROC AUC

# 4. Evaluacija
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average='binary')
rec = recall_score(y_test, y_pred, average='binary')
f1 = f1_score(y_test, y_pred, average='binary')
cm = confusion_matrix(y_test, y_pred)

# 5. Ispis metrika
print("Test skup - evaluacija:")
print(f"âœ”ï¸ TaÄnost (Accuracy): {acc:.4f}")
print(f"âœ”ï¸ Preciznost (Precision): {prec:.4f}")
print(f"âœ”ï¸ Osetljivost (Recall): {rec:.4f}")
print(f"âœ”ï¸ F1 skor: {f1:.4f}")
print("âœ”ï¸ Confusion matrix:")
print(cm)

# 6. Classification report
print("\nðŸ“Š Classification Report:")
print(classification_report(y_test, y_pred))

# 7. ROC AUC i ROC kriva
roc_auc = roc_auc_score(y_test, y_proba)
fpr, tpr, _ = roc_curve(y_test, y_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"ROC AUC = {roc_auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC AUC Curve")
plt.legend(loc="lower right")
plt.grid()
plt.tight_layout()
plt.show()

# 8. Matrica konfuzije kao heatmap
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=["Predicted 0", "Predicted 1"],
            yticklabels=["Actual 0", "Actual 1"])
plt.xlabel("Predikcija")
plt.ylabel("Stvarna vrednost")
plt.title("Matrica konfuzije")
plt.tight_layout()
plt.show()
```
