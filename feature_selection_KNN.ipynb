```python
from sklearn.pipeline import Pipeline
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.preprocessing import StandardScaler
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV, TimeSeriesSplit

# Kreiranje vremenske kros validacije
tscv = TimeSeriesSplit(n_splits=3)

# Definisanje pipeline-a
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('feature_selection', SelectKBest(score_func=f_classif)),
    ('knn', KNeighborsClassifier())
])

# Parametri za GridSearch
param_grid = {
    'feature_selection__k': [20],
    'knn__n_neighbors': [15, 25, 35]
}

# GridSearch sa TimeSeriesSplit
grid = GridSearchCV(pipe, param_grid, cv=tscv, scoring='accuracy')

# Fitovanje modela
X_filtered = X_train.drop(columns=['item_nbr', 'WeekStartDate', 'family'])
grid.fit(X_filtered, y_train)

# Ispis rezultata
print("Najbolji broj feature-a:", grid.best_params_['feature_selection__k'])
print('Najbolji broj suseda:', grid.best_params_['knn__n_neighbors'])
print("Najbolja taÄnost:", grid.best_score_)

# Prikaz izabranih feature-a
best_pipeline = grid.best_estimator_
selected_features_mask = best_pipeline.named_steps['feature_selection'].get_support()
selected_features = X_filtered.columns[selected_features_mask]
print("Odabrani feature-i:", selected_features.tolist())
```

```python
import optuna
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import TimeSeriesSplit, cross_val_score
from sklearn.pipeline import Pipeline

# âœ… X_selected treba da sadrÅ¾i veÄ‡ selektovane kolone (npr. SelectKBest iz GridSearch-a)
# Ako veÄ‡ imaÅ¡ best_pipeline, ovo veÄ‡ radi:
selected_columns = X_filtered.columns[best_pipeline.named_steps['feature_selection'].get_support()]
X_selected = X_filtered[selected_columns]

# âœ… Optuna objective funkcija za KNN
def objective(trial):
    params = {
        'n_neighbors': trial.suggest_int('n_neighbors', 3, 75),
        'weights': trial.suggest_categorical('weights', ['uniform', 'distance']),
        'p': trial.suggest_int('p', 1, 2)  # 1 = Manhattan, 2 = Euclidean
    }

    # Pipeline sa KNN (nije striktno potreban, ali dobar za buduÄ‡e proÅ¡irenje)
    pipe = Pipeline([
        ('knn', KNeighborsClassifier(**params))
    ])

    # TimeSeriesSplit kros validacija
    tscv = TimeSeriesSplit(n_splits=3)
    scores = cross_val_score(pipe, X_selected, y_train, cv=tscv, scoring='accuracy', n_jobs=-1)

    return scores.mean()

# âœ… Pokretanje optimizacije
study = optuna.create_study(direction='maximize')
study.optimize(objective, n_trials=30)

# âœ… Prikaz rezultata
print("Najbolji parametri:", study.best_params)
print("Najbolja taÄnost:", study.best_value)
```

```python
# X_selected treba da sadrÅ¾i veÄ‡ selektovane kolone (npr. SelectKBest iz GridSearch-a)
# Ako veÄ‡ imaÅ¡ best_pipeline, ovo veÄ‡ radi:
selected_columns = X_filtered.columns[best_pipeline.named_steps['feature_selection'].get_support()]
X_selected = X_filtered[selected_columns]


best_params = {
    'n_neighbors': 5,
    'weights': 'uniform',
    'p': 1
}
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix

# 1. Treniranje modela na celom trening skupu sa najboljim parametrima
final_model = KNeighborsClassifier(**best_params)
final_model.fit(X_selected, y_train)

# 2. Priprema test skupa: koristi iste kolone koje su selektovane u treningu
X_test_filtered = X_test[selected_columns]

# 3. Predikcija
y_pred = final_model.predict(X_test_filtered)

# 4. Evaluacija
acc = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average='binary')
rec = recall_score(y_test, y_pred, average='binary')
f1 = f1_score(y_test, y_pred, average='binary')
cm = confusion_matrix(y_test, y_pred)

# 5. Ispis rezultata
print("Test skup - evaluacija:")
print(f"âœ”ï¸ TaÄnost (Accuracy): {acc:.4f}")
print(f"âœ”ï¸ Preciznost (Precision): {prec:.4f}")
print(f"âœ”ï¸ Osetljivost (Recall): {rec:.4f}")
print(f"âœ”ï¸ F1 skor: {f1:.4f}")
print("âœ”ï¸ Confusion matrix:")
print(cm)
```

```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_auc_score, roc_curve, classification_report, ConfusionMatrixDisplay
import seaborn as sns


# 1. Predikcije
y_pred = final_model.predict(X_test_filtered)
y_proba = final_model.predict_proba(X_test_filtered)[:, 1]  # za ROC AUC

# 2. Evaluacija metrika
print("ðŸ“Š Classification Report:")
print(classification_report(y_test, y_pred))

# 3. ROC AUC i ROC kriva
roc_auc = roc_auc_score(y_test, y_proba)
fpr, tpr, thresholds = roc_curve(y_test, y_proba)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"ROC AUC = {roc_auc:.2f}")
plt.plot([0, 1], [0, 1], linestyle='--', color='gray')
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC AUC Curve")
plt.legend(loc="lower right")
plt.grid()
plt.tight_layout()
plt.show()

# 4. Matrica konfuzije kao heatmap
cm = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True,fmt='d', cmap='Blues', vmax=20000, 
            xticklabels=["Predicted 0", "Predicted 1"],
            yticklabels=["Actual 0", "Actual 1"]
            )
plt.xlabel("Predikcija")
plt.ylabel("Stvarna vrednost")
plt.title("Matrica konfuzije")
plt.tight_layout()
plt.show()

print(cm)
```
