```python
df_family = df
# df_family = df[(df['family'] != 'CLEANING') & (df['family'] != 'EGGS')]
print(df_family.shape)
```

```python
df_ordered_output = df_family.groupby('item_nbr')['Ordered_next_week'].sum()
#print(df_ordered_output.value_counts())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

for family in df['family'].unique():
    df_filtriran = df[df['family'] == family]
    df_grupisan = df_filtriran.groupby('item_nbr')['Ordered_next_week'].sum()

    plt.figure(figsize=(10, 6))
    sns.kdeplot(df_grupisan, fill=True, color='skyblue')

    plt.title(f'Distribucija ukupno naruÄenih jedinica za familiju: {family}')
    plt.xlabel('Ukupno naruÄeno sledeÄ‡e nedelje')
    plt.ylabel('Gustina')

    plt.grid(True)
    plt.tight_layout()
    plt.show()
```

```python
df_ordered_output.describe()
```

```python
# 2. Kvartili
q1 = df_ordered_output.quantile(0.3)
q3 = df_ordered_output.quantile(0.7)
print(q1)
print(q3)
# 3. Filtriraj item_nbr koji su izmeÄ‘u Q1 i Q3
valid_items = df_ordered_output[(df_ordered_output >= q1) & (df_ordered_output <= q3)].index
#len(valid_items)
# 4. Filtriraj originalni df
df_filtered = df[df['item_nbr'].isin(valid_items)]
print(df_filtered.shape)
```

```python
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt

# âœ… Pretvori Ordered i Ordered_next_week u binarne vrednosti (ako veÄ‡ nisu)
y_true = (df_filtered['Ordered_next_week'] > 0).astype(int)
y_pred = (df_filtered['Ordered'] > 0).astype(int)

# ðŸ§® Matrica konfuzije
cm = confusion_matrix(y_true, y_pred)
print("Matrica konfuzije:\n", cm)

# ðŸ“Š Detaljan izveÅ¡taj
print("\nKlasifikacioni izveÅ¡taj:\n")
print(classification_report(y_true, y_pred, digits=4))

# ðŸŽ¨ Vizuelizacija matrice konfuzije
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Nije naruÄeno", "NaruÄeno"])
disp.plot(cmap=plt.cm.Blues)
plt.title("Matrica konfuzije: Ordered (predikcija) vs Ordered_next_week (istina)")
plt.show()
```
